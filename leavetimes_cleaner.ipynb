```python
import pandas as pd
import os
```

```python
def clean_leavetimes(df, filename):
    print("Changing datatypes")
    categorical_cols = df.select_dtypes(['object']).columns
    categorical_cols = categorical_cols.append(df[['TRIPID', 'PROGRNUMBER', 'STOPPOINTID', 'VEHICLEID', 'JUSTIFICATIONID', 'SUPPRESSED']].columns)
    # convert columns in the list to categorical columns
    for col in categorical_cols:
        df[col] = df[col].astype('category')
    print("Dropping", str(df.duplicated().sum()), "duplicates")
    df.drop_duplicates()
    print("Dropping unnecessary columns")
    df.drop(columns=['DATASOURCE','SUPPRESSED','PLANNEDTIME_DEP','VEHICLEID','PASSENGERS','PASSENGERSIN','PASSENGERSOUT','DISTANCE','JUSTIFICATIONID','LASTUPDATE','NOTE'], inplace=True)
    print("Renaming columns")
    df.rename(columns={'PLANNEDTIME_ARR':'PLANNEDTIME_ARR_DEP'}, inplace=True)
    #save to new csv
    print("Saving cleaned dataframe as", str(f'lt_{filename}_cleaned.csv'))
    df.to_csv(f'lt_{filename}_cleaned.csv', index=False)
```

```python
for filename in range(1,25):
    f = os.path.join('/Users/rebeccadillon/git/dublin-bus-team-5/data/rt_leavetimes', f'lt-{filename}.txt')
    if os.path.isfile(f):
        print("Now cleaning", str(f))
        df = pd.read_csv(f, parse_dates=['DAYOFSERVICE', 'LASTUPDATE'], sep=';')
        clean_leavetimes(df, filename)
```
